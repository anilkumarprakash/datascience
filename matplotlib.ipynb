# Creating the plot

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60])
yvalue = np.array([15,25,35,45])
plt.plot(xvalue, yvalue, linestyle= 'dashed', color='green', marker ='o', ms=5, mec='hotpink', mfc='hotpink')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Attribute 3")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60])
yvalue = np.array([15,25,35,45])
plt.plot(xvalue, yvalue, label='Line', marker='o')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Attribute 3")
plt.legend(loc='best')  # legend() is used to add a legend to the plot, loc='best' places it in the best location
plt.show()
```

# set width of line of plot using linewidth

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60])
yvalue = np.array([15,25,35,45])
plt.plot(xvalue, yvalue, linewidth=12)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Attribute 3")
plt.show()
```

# Set color of Line and label of line

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60])
yvalue = np.array([15,25,35,38])
year = np.array([2005,2010,2015,2020])
plt.plot(year, xvalue, marker='o', color='purple', label='Value1')
plt.plot(year, yvalue, marker='o', color='green', label='Value2')
plt.xlabel("Years")
plt.ylabel("Value")
plt.title("Histogram of Attribute 3")
plt.legend(loc='best')  # legend() is used to show the label of plots
plt.show()
```

# Scatter Plot
- This plot is numerical vs numerical
- often used for finding correlation 

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60,50,60,70,80,75,95,60,85])
yvalue = np.array([15,25,35,38,40,45,50,55,60,65,70,75])
plt.scatter(xvalue, yvalue, color='purple', marker='o')
plt.xlabel("Value 1")
plt.ylabel("Value 2")           
plt.title("Scatter Plot between Value 1 and Value 2")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv')

# clean column names (remove stray quotes and surrounding whitespace)
df.columns = df.columns.str.replace('"', '', regex=False).str.strip()
```

```python
plt.scatter(df['Height(Inches)'], df['Weight(Pounds)'], color='blue', marker='o')
plt.xlabel("Height (Inches)")
plt.ylabel("Weight (Pounds)")
plt.title("Height vs Weight")
plt.show()
```

# StacBar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60,50,60,70,80,75,95,60,80])
yvalue = np.array([15,25,35,38,40,45,50,55,60,65,70,75])
year = np.array([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020,2022])
plt.bar(year, xvalue, width=1.5, color='purple', label='Value1')
plt.bar(year, yvalue , width=1.5, color='green', label='Value2', bottom=xvalue)
plt.xticks(rotation=90)
plt.xlabel("Years")
plt.ylabel("Value")
plt.title("Histogram of Attribute 3")
plt.legend(loc='best')  # legend() is used to show the label of plots
plt.show()
```
