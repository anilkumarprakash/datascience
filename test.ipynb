```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60])
yvalue = np.array([15,25,35,45])
plt.plot(xvalue, yvalue, linestyle= 'dashed', color='green', marker ='o', ms=5, mec='hotpink', mfc='hotpink')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Attribute 3")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60])
yvalue = np.array([15,25,35,45])
plt.plot(xvalue, yvalue, ls= 'dotted', c='green', marker ='o', mfc='r')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Attribute 3")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xvalue = np.array([10,20,50,60])
yvalue = np.array([15,25,35,45])
plt.plot(xvalue, yvalue, linewidth='10', c='green')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Attribute 3")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1, y1, label = "Line 1", color='r')
plt.plot(x2, y2, label = "Line 2", color='g')
plt.title("Two lines on same graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")        
```
