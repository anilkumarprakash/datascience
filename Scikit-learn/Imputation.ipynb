```python
import pandas as pd
```

```python
df = pd.read_csv("8_df.csv")
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy = 'mean')    
```

```python
df['A'] = si.fit_transform(df[['A']])
df['D'] = si.fit_transform(df[['D']])
```

```python
df
```

```python
si_mode = SimpleImputer(strategy = 'median')    
```

```python
df['B']  =  si_mode.fit_transform(df[['B']])
```

```python
df
```

```python
si_mf = SimpleImputer(strategy = 'most_frequent')    
```

```python
df['C'] = si_mf.fit_transform(df[['C']])
```
