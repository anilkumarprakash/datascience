# One Hot Encoder and get Dummies

```python
import pandas as pd
```

```python
df = pd.read_excel('onedata.xlsx')
```

```python
df
```

```python
pd.get_dummies(data=df,columns=['buyer','fruits','gender'],dtype=int)
```

```python
pd.get_dummies(data=df,columns=['buyer','fruits','gender'],dtype=int, drop_first='first')
```

# OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', dtype=int, sparse_output=False)
```

```python
df_transf = ohe.fit_transform(df[['buyer','fruits','gender']])
```

```python
df_transf
```

# This function is used for getting column name after transform

get_feature_names_out()

```python
cols = ohe.get_feature_names_out()
```

```python
cols
```

```python
df_transf = pd.DataFrame(data = df_transf, columns = cols)
```
