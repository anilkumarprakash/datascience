# Seaborn
Types of function
- Figure level
- Axes levels

Types of Plots
- Relational plot (scatter, Line)
- Distribution plot (Histogram, kcd, regplot)
- Categorical Plot (bar, count, box, voilin, swam)
- Matrix plot (heatmap, clustermap)
- Regression plot
- Multi plot (joint, pair)


```python
%pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = sns.load_dataset("tips")
data.head(50)
```

Scatterplot

```python
sns.scatterplot(x='total_bill', y='tip', data=data)
```

Relplot
- relplot can used instead of scatterplot

```python
sns.relplot(x='total_bill', y='tip', kind='scatter', data=data)
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=data, hue='smoker', style='time')
```

```python
sns.relplot(x='total_bill', y='tip', data=data, hue='smoker', style='time', size='size', kind='scatter')
```

```python
sns.relplot(x='total_bill', y='tip', data=data, col='smoker', kind='scatter')
```

Lineplot

```python
x = np.array([2, 4, 6 , 8, 10])
y = np.array([21, 12, 33, 44, 50])
sns.lineplot(x=x, y=y)
```

```python
sns.relplot(x=x,y=y, kind='line')
```

# Distribution Plots
- histplot
- kdeplot
- ecdfplot
- displot
- jointplot

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.histplot(x='total_bill', data=tips)
```

```python
sns.displot(tips['day'], kind='hist')
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.displot(x='survived', kind='hist', data=titanic , element='step' )
```

# kdeplot

```python
sns.kdeplot(x='age', data=titanic, fill=True, hue='sex')
```

```python
sns.displot(tips['total_bill'], kind='kde', fill=True)
```

```python
sns.displot(tips['total_bill'], kde=True)
```

# rugplot

```python
sns.displot(tips['total_bill'], kind='kde', fill=True)
sns.rugplot(x='total_bill', data=tips)
```

# histmap
- it is a axes level function

```python
temp = tips.pivot_table(index='day', columns='time', values='total_bill', aggfunc='mean')
sns.heatmap(temp, annot=True, fmt=".1f", cmap='YlGnBu', linewidths=0.5)  

# linewidths: width of lines that will divide each cell
# annot: annotation on cells
# fmt: format of annotation (e.g., .1f for one decimal places)
# cmap: color map
plt.figure(figsize=(8, 6))
```
